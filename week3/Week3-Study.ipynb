```python
def calc_mean_review_scores(group):
    avg = np.nanmean(group['review_scores_value'])
    
    group['review_scores_value'] = np.abs(avg - group['review_scores_value'])
    
    return group

df.groupby('cancellation_policy').apply(calc_mean_review_scores).head()
```

# Pivot Table

```python
df = pd.read_csv('resources/week-3/datasets/cwurData.csv')
df.head()
```

```python
def create_category(ranking):
    if (ranking >= 1) & (ranking <= 100):
        return "First Top Tier University"
    elif (ranking >= 101) & (ranking <= 200):
        return "Secpnd Top Tier University"    
    elif (ranking >= 201) & (ranking <= 300):
        return "Third Top Tier University"
    return "Other Top University"

df['Rank_Level'] = df['world_rank'].apply(lambda df: create_category(df))

df.head()
```

```python
df.pivot_table(values = 'score'
               , index = 'country'
               , columns = 'Rank_Level'
               , aggfunc = [np.mean]).head()
```

```python
df.pivot_table(
    values = 'score', 
    index = 'country', 
    columns = 'Rank_Level', 
    aggfunc = [np.mean, np.max]).head()
```

```python
df.pivot_table(values = 'score'
               , index = 'country'
               , columns = 'Rank_Level'
               , aggfunc = [np.mean, np.max],
               margins = True).head()
```

```python
df.pivot_table(values = 'score'
               , index = 'country'
               , columns = 'Rank_Level'
               , aggfunc = [np.mean, np.max]
              ).head()
```

```python
new_df = df.pivot_table(values = 'score'
                        , index = 'country'
                        , columns = 'Rank_Level'
                        , aggfunc = [np.mean, np.max],
               margins = True)

print(new_df.index)
print(new_df.columns)
```
